```python
import numpy as np
import matplotlib.pyplot as plt

# sketch x^3-5x+3
x = np.linspace(-5, 5, int(1e5))
f = lambda x: x**3 - 5* x + 3
fig, ax = plt.subplots(1,1)
ax.plot(x, f(x), color='steelblue')
ax.plot(x, 0*x, color='orange', ls=':')
ax.set_xlabel('x')

ax.set_ylabel('function')
ax.set_title('sketch')
plt.show()
```
