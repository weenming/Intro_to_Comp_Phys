```python
import numpy as np
import matplotlib.pyplot as plt

sol_e = np.loadtxt('solution_explicit.txt', dtype='complex')
portions_e = np.loadtxt('portions_explicit.txt', dtype='float')

sol_CN = np.loadtxt('solution_CN.txt', dtype='complex')
portions_CN = np.loadtxt('portions_CN.txt', dtype='float')

t_max = 20
fig, ax = plt.subplots(1, 1)
ax.plot(np.linspace(0, t_max, len(portions_e)),
        [p[0] for p in portions_e], label='explicit left')
ax.plot(np.linspace(0, t_max, len(portions_e)),
        [p[1] for p in portions_e], label='explicit right')
ax.plot(np.linspace(0, t_max, len(portions_e)),
        [p[2] for p in portions_e], label='explicit inside')

ax.plot(np.linspace(0, t_max, len(portions_CN)),
        [p[0] for p in portions_CN], label='CN left', ls='--')
ax.plot(np.linspace(0, t_max, len(portions_CN)),
        [p[1] for p in portions_CN], label='CN right', ls='--')
ax.plot(np.linspace(0, t_max, len(portions_CN)),
        [p[2] for p in portions_CN], label='CN inside', ls='--')
ax.set_xlabel('time / s')
ax.set_ylabel('portion')
ax.legend()
plt.show()
```

```python
x = np.linspace(-30, 30, 1000)

fig, ax = plt.subplots(1, 2, sharey=True)
fig.set_size_inches(7, 4)
s = ax[0].imshow(np.abs(sol_e) ** 2, extent=[x[0], x[-1], t_max, 0], aspect=5)
ax[0].set_ylim(ax[0].get_ylim()[::-1])
ax[0].set_ylabel('time / s')
ax[0].set_xlabel('x / a.u.')
ax[0].set_title('Stable explicit scheme')
s = ax[1].imshow(np.abs(sol_CN) ** 2, extent=[x[0], x[-1], t_max, 0], aspect=5)
# ax[1].set_ylim(ax[1].get_ylim()[::-1])
ax[1].set_title('Crank-Nicolson scheme')
ax[1].set_xlabel('x / a.u.')
fig.colorbar(s)
```
