```python
import numpy as np
import matplotlib.pyplot as plt

def v_of_sphere(dim, N = int(1e4)):
    c = 0
    for _ in range(N):
        x = np.random.random(dim)
        if np.sum(np.abs(x) ** 2) < 1:
            c += 1
    return c / N * 2 ** dim

def v_of_sphere_theory(n):
    if n in [0, 1]:
        return (1, 2)[n]
    else:
        return 2 * np.pi / n * v_of_sphere_theory(n - 2)

print('2:', v_of_sphere(2), np.pi)

print('3:', v_of_sphere(3), 4 * np.pi / 3)

print('4:', v_of_sphere(4), v_of_sphere_theory(4))
print('5:', v_of_sphere(5), v_of_sphere_theory(5))


fig, ax = plt.subplots(1, 1)
ns = np.arange(15)
ax.plot(ns, [v_of_sphere(n, 10 ** (5)) for n in ns], label='Monte Carlo, hit and miss')
ax.plot(ns, [v_of_sphere_theory(n) for n in ns], label='Theory')
ax.legend()
ax.set_xlabel('Dimension')
ax.set_ylabel('Volume')
fig.savefig('q1.png', dpu=300)
```
